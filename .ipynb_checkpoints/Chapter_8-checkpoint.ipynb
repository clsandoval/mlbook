```python
import sklearn
import numpy as np
from matplotlib import cm
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### PCA

```python
#3d dataset
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)
```

```python
X.shape
```

```python
X_centered = X - X.mean(axis = 0)
U, s , Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:,0]
c2 = Vt.T[:,1]
```

```python
### principal unit vectors , in order of importance
print(c1,c2)
```

### Projecting to d dimensions
$$X_{d-proj} = XW_{d}$$

```python
W_2 = Vt.T[:,:2]
X_2D = X_centered.dot(W_2)
X_2D.shape
```

### PCA with Scikit Learn

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
X_2D = pca.fit_transform(X)
print(X_2D.shape)
#2 principal component unit vectors
print(pca.components_.T[:,:2])
#Variance Ratio of each component
print(pca.explained_variance_ratio_)
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml("mnist_784", version = 1, as_frame = False)
mnist.target = (mnist.target.astype(int))
```

```python
from sklearn.model_selection import train_test_split
X = mnist['data']
y = mnist['target']

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
pca = PCA(n_components = 0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
X_recovered = pca.inverse_transform(X_reduced)
```

```python
def plot_digits(ax,num, data, w,h):
    grid = np.array([])
    grid2 = []
    for i in range(num,num**2,num):
        sample_recovered = data[i-5:i]
        grid = np.concatenate(sample_recovered.reshape(-1,w,h),axis = 0)
        grid2.append(grid)
    grid2 = np.array(grid2)
    print(grid2.shape)
    grid_final=np.concatenate(grid2,axis = 1)
    ax.imshow(grid_final)
fig,ax = plt.subplots(nrows=1, ncols=2)
plot_digits(ax[0],5, X_train, 28,28)
plot_digits(ax[1],5, X_recovered, 28,28)
```

### Randomized PCA

```python
rnd_pca = PCA(n_components = 154, svd_solver = "randomized")
X_reduced = rnd_pca.fit_transform(X_train)
```

```python
from sklearn.decomposition import IncrementalPCA
n_batches = 100
inc_pca = IncrementalPCA(n_components = 154)
for X_batch in np.array_split(X_train,n_batches):
    inc_pca.partial_fit(X_batch)
X_reduced = inc_pca.transform(X_train)
```

#### Load MNIST and train a random forest classifier with and w/o PCA

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

mnist = fetch_openml("mnist_784", version = 1, as_frame = False)
mnist.target = (mnist.target.astype(int))
X = mnist['data']
y = mnist['target']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
rnd_forest_no_pca = RandomForestClassifier(verbose = 1, n_jobs = -1)
rnd_forest_no_pca.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
preds = rnd_forest_no_pca.predict(X_test)
print(accuracy_score(preds,y_test))
```

```python
pca = PCA(n_components = 0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
rnd_forest_pca = RandomForestClassifier(verbose = 1, n_jobs = -1)
rnd_forest_pca.fit(X_reduced,y_train)
```

```python
X_test_reduced = pca.transform(X_test)
preds = rnd_forest_pca.predict(X_test_reduced)
print(accuracy_score(preds,y_test))
```

#### Use TSNE to reduce MNIST to two dimensions, plot with matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
X_subsample = X[:10000]
y_subsample = y[:10000]
tsne = TSNE(n_components = 3)
X_subsample_reduced = tsne.fit_transform(X_subsample)
```

```python
plt.figure(figsize=(13,10))
plt.scatter(X_subsample_reduced[:,0], X_subsample_reduced[:,1], c = y_subsample, cmap = 'jet')
plt.axis('off')
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (13,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(*zip(*X_subsample_reduced), c = y_subsample, cmap = 'jet')
plt.show()
```

#### PCA -> TSNE

```python
from sklearn.pipeline import Pipeline
pca_tsne_pipe = Pipeline([
    ('pca', PCA(n_components = 0.95)),
    ('tsne', TSNE(n_components = 2))
])
```

```python
X_subsample_reduced = pca_tsne_pipe.fit_transform(X_subsample)
```

```python
plt.figure(figsize=(13,10))
plt.scatter(X_subsample_reduced[:,0], X_subsample_reduced[:,1], c = y_subsample, cmap = 'jet')
plt.axis('off')
plt.colorbar()
plt.show()
```
